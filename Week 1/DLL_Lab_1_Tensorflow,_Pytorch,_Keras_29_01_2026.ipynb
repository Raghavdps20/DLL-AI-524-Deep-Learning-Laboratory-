```python
# TensorFlow (Keras)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Build the model
model = Sequential([
    Dense(256, activation='relu', input_shape=(64,)),
    Dense(10)   # output layer
])

# Compile the model
model.compile(
    optimizer=Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

# Dummy training data (for Colab demo)
X_train = tf.random.normal((1000, 64))
y_train = tf.random.uniform((1000,), minval=0, maxval=10, dtype=tf.int32)

# Train the model
model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32
)

# Save the model
model.save("/content/my_model.keras")

from tensorflow.keras.models import load_model

loaded_model = load_model("/content/my_model.keras")
print("Model loaded successfully!")


```

```python
# PyTorch
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Define model
model = nn.Sequential(
    nn.Linear(64, 256),
    nn.ReLU(),
    nn.Linear(256, 10)
)

# Loss & optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())

# Dummy training data
X_train = torch.randn(1000, 64)
y_train = torch.randint(0, 10, (1000,))

dataset = TensorDataset(X_train, y_train)
loader = DataLoader(dataset, batch_size=32, shuffle=True)

# Training loop
for epoch in range(10):
    for x_batch, y_batch in loader:
        optimizer.zero_grad()
        outputs = model(x_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}/10 completed")

# Save model
torch.save(model.state_dict(), "my_model.pth")

# Load model
loaded_model = nn.Sequential(
    nn.Linear(64, 256),
    nn.ReLU(),
    nn.Linear(256, 10)
)
loaded_model.load_state_dict(torch.load("my_model.pth"))
loaded_model.eval()

print("Model saved and loaded successfully!")
```

```python
# 1) Define a 2 layer sequential model, compile it, & run for 30 epochs, & print the predicted results.
# X1 = [1,2,3,4]
# Y1 = [2,4,6,8]
# Y = 2X
# Linear model
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Data
X1 = np.array([1, 2, 3, 4])
Y1 = np.array([2, 4, 6, 8])

# Define a 2-layer Sequential model
model = Sequential([
    Dense(1, input_shape=(1,), activation='linear'),  # Layer 1
    Dense(1, activation='linear')                      # Layer 2
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)

# Train the model
model.fit(X1, Y1, epochs=30, verbose=0)

# Predict
predictions = model.predict(X1)

# Print results
print("Input X:", X1)
print("Predicted Y:", predictions.flatten())
```
